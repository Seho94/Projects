```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
keyword = input()
driver = webdriver.Chrome()
url = "https://www.youtube.com/results?search_query="+str(keyword)
# url = "https://www.python.org"
driver.get(url)
html = driver.page_source
soup = BeautifulSoup(html)
```

```python
# 단일 영상 정보
basic = soup.find_all("div",{"id":"contents", "class":"style-scope ytd-item-section-renderer"})[-1]
video = basic.find_all("a", {"id":"video-title"})

# 동영상 url
link = "https://wwww.youtube.com" + video[0]['href']

# 채널명
channel = basic.find_all('div',{'id':'channel-info'})[2] # 인덱스: 채널명 종류
channel = channel.find_all("yt-formatted-string")
channel = channel[0].text # 0 고정

channel_list = []
# 채널리스트
for item in basic.find_all('div',{'id':'channel-info'}):
    item = item.find_all("yt-formatted-string")
    item = item[0].text
    channel_list.append(item)

title_list = []
num_list = []
date_list = []
link_list = []

for i in range(len(channel_list)):
    link = "https://wwww.youtube.com" + video[i]['href']
    tempData = video[i]['aria-label']
    tempData = tempData.split(' 게')
    title= tempData.pop(0) # 제목
    tempData = tempData[0]
    tempData = tempData.split(' ')
    num = tempData.pop() # 조회수
    tempData.pop() # '조회수' 삭제
    tempData.pop(0) # '시자:'' 삭제
    
    length_channel = len(channel_list[i].split(" ")) # 채널이름에 띄어쓰기가 존재하는 경우를 위함.
    date = ' '.join(tempData[length_channel:]) # 업로드 일자

    title_list.append(title)
    link_list.append(link)
    num_list.append(num)
    date_list.append(date)
    
df = pd.DataFrame({'제목':title_list,
                  '채널명':channel_list,
                  '업로드일자':date_list,
                  '조회수':num_list,
                  'link':link_list})
df
```

```python
# 플레이리스트 영상 정보
basic = soup.find_all("div",{"id":"contents", "class":"style-scope ytd-item-section-renderer"})[-1]
video = basic.find_all("ytd-playlist-renderer")
title = video[0].find_all("span", {"id":"video-title"})[0]['title'] # 재생목록 제목
channel = video[0].find_all("div",{"id":"tooltip"})[0].text.replace("\n", "").strip() # 채널명
title_1 = video[0].find_all("ytd-child-video-renderer")[0].find_all("a")[0].find_all("yt-formatted-string")[0]["title"] # 재생목록 하위영상 제목
title_2 = video[0].find_all("ytd-child-video-renderer")[1].find_all("a")[0].find_all("yt-formatted-string")[0]["title"] # 재생목록 하위영상 제목
num1 = video[0].find_all("a", {"id":"thumbnail"})[0].find_all("ytd-thumbnail-overlay-side-panel-renderer")[0].text.replace("\n", "") # 1번
num2 = video[1].find_all("a", {"id":"thumbnail"})[0].find_all("ytd-thumbnail-overlay-side-panel-renderer")[0].text.replace("\n", "") # 2번
link1 = video[0].find_all("ytd-child-video-renderer")[0].find_all("a")[0]["href"] # 1번링크
link2 = video[0].find_all("ytd-child-video-renderer")[1].find_all("a")[0]["href"] # 2번링크

titleList, channelList, titles, numList, links = [],[],[],[],[]

for i in range(len(video)):
    tempTitle, tempLink = [], []
    title = video[i].find_all("span", {"id":"video-title"})[0]['title'] # 재생목록 제목
    channel = video[i].find_all("div",{"id":"tooltip"})[0].text.replace("\n", "").strip() # 채널명
    num = video[i].find_all("a", {"id":"thumbnail"})[0].find_all("ytd-thumbnail-overlay-side-panel-renderer")[0].text.replace("\n", "") # 1번
    for j in range(len(video[0].find_all("ytd-child-video-renderer"))):
        title1 = video[i].find_all("ytd-child-video-renderer")[j].find_all("a")[0].find_all("yt-formatted-string")[0]["title"] # 재생목록 하위영상 제목
        link1 = "https://www.youtube.com" + video[i].find_all("ytd-child-video-renderer")[j].find_all("a")[0]["href"] # 1번링크
        tempTitle.append(title1)
        tempLink.append(link1)
    titleList.append(title)
    channelList.append(channel)
    numList.append(num)
    titles.append(tempTitle)
    links.append(tempLink)

# 제생목록, 채널명, 영상 수, 하위제목1, 하위링크1, 하위제목2, 하위링크2
df = pd.DataFrame({"재생목록":titleList,
                  "채널명":channelList,
                  "영상 수":numList})


for j in range(len(titles[0])):
    df["하위제목"+str(j+1)] = None
    df["하위링크"+str(j+1)] = None

for i in range(len(df)):
    for j in range(len(titles[0])):
        df["하위제목"+str(j+1)][i] = titles[i][j]
        df["하위링크"+str(j+1)][i] = links[i][j]

df
```
